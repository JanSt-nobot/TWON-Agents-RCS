```python
import json

import pandas as pd
import seaborn as sns
```

```python
DATA_PATH: str = './annotated/dataset.german.json'
```

```python
dataset: pd.DataFrame = (
    pd.json_normalize(json.load(open(DATA_PATH, 'r')))
    .set_index('id')
    .astype({
        'persona': 'category', 'model': 'category',
        'topic': 'category', 'platform': 'category',
        'language': 'category'
    })
)
dataset
```

```python
dataset.groupby('model', observed=False).describe(percentiles=[]).T
```

```python
sns.violinplot(data=dataset, x="model", y="annotation.authenticity", inner="quart", split=True)
```

```python
dataset.groupby(['persona', 'model'], observed=False).describe(percentiles=[]).T
```

```python
sns.violinplot(data=dataset, x="persona", y="annotation.persona", hue="model", split=True, inner="quart")
```

```python
dataset.groupby(['topic', 'model'], observed=False).describe(percentiles=[]).T
```

```python
sns.violinplot(data=dataset, x="topic", y="annotation.topic", hue="model", split=True, inner="quart")
```

```python
dataset.groupby(['platform', 'model'], observed=False).describe(percentiles=[]).T
```

```python
sns.violinplot(data=dataset, x="platform", y="annotation.authenticity", hue="model", split=True, inner="quart")
```
