```python
import json

import pandas as pd
```

```python
DATA_PATH: str = './annotated/dataset.german.json'
```

```python
dataset: pd.DataFrame = (
    pd.json_normalize(json.load(open(DATA_PATH, 'r')))
    .set_index('id')
    .astype({
        'persona': 'category', 'model': 'category',
        'topic': 'category', 'platform': 'category',
        'language': 'category'
    })
)
dataset
```

```python
dataset.groupby('model', observed=False).describe(percentiles=[]).T
```

```python
dataset.groupby(['persona', 'model'], observed=False).describe(percentiles=[]).T
```

```python
dataset.groupby(['topic', 'model'], observed=False).describe(percentiles=[]).T
```

```python
dataset.groupby(['platform', 'model'], observed=False).describe(percentiles=[]).T
```
